### Build operational views - success and exceptions

```sql
CREATE OR REPLACE VIEW payment_app.gold.v_success_transactions AS
SELECT
    f.txn_id,
    f.date_key,
    b.biller_name,
    f.txn_source,
    f.txn_amount,
    f.credited_amount,
    f.load_ts
FROM payment_app.gold.fact_reconciliation f
LEFT JOIN payment_app.gold.dim_biller b
    ON f.biller_key = b.biller_key
WHERE f.recon_flag = 'Y';
```

```sql
CREATE OR REPLACE VIEW payment_app.gold.v_exception_transactions AS
SELECT
    f.txn_id,
    f.date_key,
    b.biller_name,
    f.txn_source,
    f.txn_amount,
    f.credited_amount,
    f.load_ts
FROM payment_app.gold.fact_reconciliation f
LEFT JOIN payment_app.gold.dim_biller b
    ON f.biller_key = b.biller_key
WHERE f.recon_flag = 'N';
```
