## # Set-up Project environment

 1.Create storage credentials
 2.Create external location

1.Create storage credential in UI using access connector (after providing storage blob data contributor access to datalake)

```sql
--CREATE EXTERNAL LOCATION 

CREATE EXTERNAL LOCATION IF NOT EXISTS ext_payment_app
URL 'abfss://payment-app@bankingdl.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `project-credential`)
```

```sql
--CREATE CATALOG

CREATE CATALOG IF NOT EXISTS payment_app
MANAGED LOCATION 'abfss://payment-app@bankingdl.dfs.core.windows.net/'
```

```sql
CREATE SCHEMA IF NOT EXISTS payment_app.landing
CREATE SCHEMA IF NOT EXISTS payment_app.bronze
MANAGED LOCATION 'abfss://payment-app@bankingdl.dfs.core.windows.net/bronze'
CREATE SCHEMA IF NOT EXISTS payment_app.silver
MANAGED LOCATION 'abfss://payment-app@bankingdl.dfs.core.windows.net/silver'
CREATE SCHEMA IF NOT EXISTS payment_app.gold
MANAGED LOCATION 'abfss://payment-app@bankingdl.dfs.core.windows.net/gold'
```

```sql
CREATE EXTERNAL VOLUME payment_app.landing.files
LOCATION 'abfss://payment-app@bankingdl.dfs.core.windows.net/landing'
```
